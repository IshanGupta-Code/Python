```python
df_country = pd.read_excel('./Zomatodataset/Country-Code.xlsx')
print(df_country)
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df.head(2)
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val.tolist()[:3], labels=country_name.tolist()[:3], autopct="%1.2f%%")
```

Observation: Zomato maximum transaction records form India, then US then UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
rating_pie = ratings.groupby('Rating text').sum().reset_index()
```

```python
rating_index = rating_pie['Rating text'].value_counts().index
rating_index
```

```python
rating_val = rating_pie['Rating count'].values
rating_val
```

```python
plt.pie(rating_val.tolist(), labels=rating_index.tolist(), autopct="%1.2f%%")
```
