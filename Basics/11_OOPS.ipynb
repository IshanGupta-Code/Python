# **Proj 1 (Banking System)**

```python
class bank_account:
    def __init__(self, bal, acc):
        self.bal = bal
        self.account_no = acc

    # Debit
    def debit(self, amount):
        self.bal -= amount
        print('Rs.', amount, 'is debited')
        print('Remaining Balance Rs.', self.get_balance())

    # Credit
    def credit(self, amount):
        self.bal += amount
        print('Rs.', amount, 'is credited')
        print('Total Balance Rs.', self.get_balance())

    # Get Balance
    def get_balance(self):
        return self.bal

acc1 = bank_account(10000, 6470)
acc1.debit(1000)
acc1.credit(10000)
acc1.debit(5000)
```

## **Inheritence**

### Single Inheritence


```python
# When one class(child/derived) drives properties and methods of another class(Parent class)

# Type1 :- Single Inheritence
class Car:
    color = 'black'
    @staticmethod
    def start():
        print('starting...')

    @staticmethod
    def stop():
        print('stop.')

# toyota_car (Child Class) inherits the properties of Car (Parent Class)
class toyota_car(Car):
    def __init__(self, name):
        self.name = name

car1 = toyota_car('Fortuner')
car2 = toyota_car('Pirus')

car1.start()
print(car1.name)
```

### Multi-Level Inheritence

```python
# Type1 :- Single Inheritence
class Car:
    color = 'black'
    @staticmethod
    def start():
        print('starting...')

    @staticmethod
    def stop():
        print('stop.')

# toyota_car (Child Class) inherits the properties of Car (Parent Class)
class toyota_car(Car):
    def brand(self, brand):
        self.brand = brand

# Fortuner (Child Class) inherits the properties of both toyota_car (Parent Class) and Car (Parent Class)
class Fortuner(toyota_car):
    def __init__(self, type):
        self.type = type

car1 = Fortuner('Petrol')
car1 = Fortuner('Disel')

car1.start()
print(car1.type)

```

### Multiple Inheritence

```python
class A:
    varA = "Welcome to class A"

class B:
    varB = "Welcome to Class B"

class C(A,B):
    varC = "Welcome to Class C"

stu1 = C()

print(stu1.varA)
print(stu1.varB)
print(stu1.varC)
```

```python
# Type1 :- Single Inheritence
class Car:
    def __init__(self, type):
        self.type = type

    @staticmethod
    def start():
        print('starting...')

    @staticmethod
    def stop():
        print('stop.')

# toyota_car (Child Class) inherits the properties of Car (Parent Class)
class toyota_car(Car):
    def __init__(self, name, type):
        super().__init__(type)
        self.name = name
        super().start()

car1 = toyota_car('Fortuner', 'Petrol')

print(car1.type)
print(car1.name)
```

### Property Decorator


```python
class Student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math
    
    @property
    def percentage(self):
        return (self.phy + self.chem + self.math)/3 

stu1 = Student(98,87,90)
print(round((stu1.percentage), 2))

stu1.chem = 99
print(round((stu1.percentage), 2))
```

### Dunder Function


```python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img
    
    def showNumber(self):
        print(self.real,'j + ', self.img,'i')

    def __add__(self, sec):     # __add__ (It is a Dunder Function)
        newReal = self.real + sec.real
        newImg = self.img + sec.img
        print(newReal,'j + ', newImg,'i')



num1 = Complex(2,3)
num2 = Complex(6,7)
num1.showNumber()
num2.showNumber()

num3 = num1 + num2
print(num3)
```

#### Circle

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def perimeter(self):
        prem = (2*3.14) * self.radius
        return print('The perimeter of the circle', prem)
    
    def area(self):
        area = 3.14 * (self.radius ** 2)
        return print('The area of the circle', area)

r1 = Circle(6)

print('The radius of circle:', r1.radius)
r1.area()
r1.perimeter()

```

#### Employee and Engineer (Class Inheritence)

```python
class Employee:
    def __init__(self, role, department, salary):
        self.role = role
        self.department = department
        self.salary = salary

    def showDetails(self):
        print("Role of Employee:", self.role)
        print("Department of Employee:", self.department)
        print("Salary of Employee:", self.salary)
    
class Enginner(Employee):
    def __init__(self, role, department, salary, name, age):
        self.name = name
        self.age = age
        super().__init__(role, department, salary)

    def showDetails(self):
        super().showDetails()
        print('Name of Employee:', self.name)
        print('Age of Employee:', self.age)



e1 = Enginner('Ml_Engineer', 'Tech', 45000, 'Ishan', '21')
e1.showDetails()
```

#### Dendur Operator

```python
class Order:
    def __init__(self, item, price):
        self.item = item
        self.price= price
    
    def __gt__(self, odr2):
        return self.price > odr2.price
    
odr1 = Order('Soap', 200000)
odr2 = Order('Ring', 50000)

print(odr1 > odr2)
```
