# **Functions**

```python
# plt.plot(x, y, color='color_name', linestyle='line_style', linewidth=value, marker='marker_symbol', label='label_name')
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1800, 1500, 2000, 2200]
# plot
plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label='2025 Sales Data')

# xlabel
plt.xlabel('Months')

# ylabel
plt.ylabel('Sales per Month')

# title
plt.title('Monthly Sales Data Report')

# legend
plt.legend(loc='upper left', fontsize=12)

#grid
plt.grid(color='pink', linestyle=':', linewidth=1)

# xlim
plt.xlim(1,4)

# ylim
plt.ylim(1000, 2500)

# xticks & ytickcs
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4'])

plt.show()
```

# Data Visualization Tools

## **Barchart**

```python
# plt.bar(x, height, color='color_name', label='label_name', width=value)
import matplotlib.pyplot as plt

product=['A', 'B', 'C', 'D']
sales=[2200,1500,2000,1400]

# plt.barh [For horizontal barchart]
plt.bar(product, sales, color='blue',  label='Sales 2025')
plt.legend()
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

## **PieChart**

```python
# plt.pie(values, labels=label_list, colors=color_name, autopct='%1.1f%%')

import matplotlib.pyplot as plt

regions=['North', 'South', 'East', 'West']
revenue=[3000,2000,2500,1500]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightgreen', 'coral'])
# plt.legend()
plt.title('Revenue as per Regions')
plt.show()
```

## **Histogram**

```python
# plt.hist(data, bins=number_of_bins, color='colorname', edgecolor='black')
import matplotlib.pyplot as plt

marks= [67,87,69,79,76,97,57,86,47,95,68,90,46,53,83,68,93,73,81]

plt.hist(marks, bins=5, color='purple', edgecolor='black')
plt.title('Score Distribution of Students')
plt.xlabel('Marks')
plt.ylabel('No. of Students') 
plt.show()
```

## **Scatter Plots**

```python
# plt.scatter(x, y, color='color_name', marker='marker_style', label='labelname')
import matplotlib.pyplot as plt

hours_studies=[1,2,3,4,5,6,7,8]
marks=[50,55,60,65,70,75,80,85]

plt.scatter(hours_studies, marks, color='darkgreen', marker='x', label='Students_Data')
# markers= x, o, ^, d, s, +
plt.xlabel('Hours Studies')
plt.ylabel('Marks')
plt.title('Relation Between Study Time and Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# plt.scatter(x, y, color='color_name', marker='marker_style', label='labelname')
import matplotlib.pyplot as plt

plt.scatter([1,2,3], [50,65,70], color='blue', label='Class A') # Group 1
plt.scatter([1,2,3], [40,60,65], color='green', label='Class B') # Group 2
# markers= x, o, ^, d, s, +
plt.xlabel('Hours Studies')
plt.ylabel('Marks')
plt.title('Relation Between 2 Class')
plt.legend()
plt.grid(True)
plt.show()
```

## **Subplots**

```python
# plt.Subplot(nrows , ncols, index)
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[25,20,30,15]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

# fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax= plt.subplots(1,2, figsize=(10,5))


x=[1,2,3,4]
y=[25,20,30,15]

ax[0].plot(x,y)
ax[0].set_title('Line Chart')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.suptitle('Comparision of Line and Bar Chart')
plt.tight_layout()


plt.show()
```
