#  1. **Two Sum**

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target:int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i , j]
        return []  # Return an empty list if no solution is found
            
```

# 2. **Median of 2 Sorted Array**

```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        concat_list = nums1 + nums2
        concat_list.sort()
        n = len(concat_list)
        mid = n // 2
        if n % 2 == 0:
            median = (concat_list[mid - 1] + concat_list[mid]) / 2
        else:
            median = concat_list[mid]
        return float(median)
```

```python
from typing import List
class Solution:
    def maxArea(self, height: List[int]) -> int:
        for i in range(len(height)):
            if max(height) > height[i]:
                s_h = height[i]
                area = s_h ** 2
        return int(area)
    
sol = Solution()
ans = sol.maxArea([1,2,1])
print(ans)
```

```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        for i in range(len(strs)):
            for j in range(len(strs)):
                if strs[i][j] == strs[i][j]:
                    print(strs[i][j])
        return ""
    
sol = Solution()
ans = sol.longestCommonPrefix(["flower","flow","flight"])
print(ans)
```
